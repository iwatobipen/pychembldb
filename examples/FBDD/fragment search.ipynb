```python
from pychembldb import *
```

```python
for a in chembldb.query(Assay).filter(Assay.description.like("% kinase %")):
    acts = []
    for act in a.activities:
        if act.compound.molecule.property is not None and act.compound.molecule.property.mw_freebase is not None and act.compound.molecule.property.mw_freebase < 250:
            acts.append(act.compound.molecule)

    if len(acts) > 20:
        print(a.chembl_id)
        print(a.description, "\n")
        for m in acts:
            print(m.chembl_id, m.structure.canonical_smiles)
        print()
```
